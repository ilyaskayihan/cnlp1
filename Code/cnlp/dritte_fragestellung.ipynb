```python
import pandas as pd
import time
from openai import OpenAI

client = OpenAI(
    api_key="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",
    base_url="https://openrouter.ai/api/v1"
)

criteria = [
    "Faktentreue (√úbereinstimmung mit Produktinformationen)",
    "Klarheit und Verst√§ndlichkeit",
    "Pr√§gnanz (Knappheit bei Erhaltung der Aussagekraft)",
    "Relevanz der Inhalte",
    "Sprachliche Qualit√§t (Grammatik, Stil)"
]

def generate_evaluation_prompt(product_facts, model_output):
    criteria_text = "\n".join([f"{i+1}. {c} (1-5 Punkte)" for i, c in enumerate(criteria)])
    return f"""
Du bist ein Experte f√ºr Produkttextevaluation. Bewerte die folgende Produktzusammenfassung anhand dieser Kriterien auf einer Skala von 1 (schlecht) bis 5 (sehr gut):

{criteria_text}

### Produktinformationen:
{product_facts}

### Produktzusammenfassung:
{model_output}

Bitte gib nur eine durch Komma getrennte Liste von 5 Zahlen zur√ºck, die die Punktzahl f√ºr jedes Kriterium darstellen, z.B.:
5,4,4,3,5
"""

def run_evaluation_prompt(prompt):
    response = client.chat.completions.create(
        model="openai/gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0,
        max_tokens=50
    )
    return response.choices[0].message.content.strip()

def main():
    # Lade die drei CSVs mit prompt1, prompt2, prompt3
    df1 = pd.read_csv("/Users/ilyas/Desktop/FHNW/4. Semester/Challenge/prompt1_dritte_fs.csv", sep=";", quotechar='"', on_bad_lines='skip', encoding="utf-8")
    df2 = pd.read_csv("/Users/ilyas/Desktop/FHNW/4. Semester/Challenge/prompt2_dritte_fs.csv", sep=";", quotechar='"', on_bad_lines='skip', encoding="utf-8")
    df3 = pd.read_csv("/Users/ilyas/Desktop/FHNW/4. Semester/Challenge/prompt3_dritte_fs.csv", sep=";", quotechar='"', on_bad_lines='skip', encoding="utf-8")

    # Alle ProductIds nehmen, die in allen drei Dateien vorkommen
    product_ids = set(df1["ProductId"]) & set(df2["ProductId"]) & set(df3["ProductId"])
    product_ids = sorted(list(product_ids))

    all_results = []

    for pid in product_ids:
        # Produktinfos (z.B. von prompt1) holen - sollte bei allen gleich sein
        product_facts = df1.loc[df1["ProductId"] == pid, "Prompt"].values[0]

        # Outputs von allen drei Prompts holen
        out1 = df1.loc[df1["ProductId"] == pid, "output_gpt-4"].values[0]
        out2 = df2.loc[df2["ProductId"] == pid, "output_gpt-4"].values[0]
        out3 = df3.loc[df3["ProductId"] == pid, "output_gpt-4"].values[0]

        # Evaluationen sammeln
        scores_per_prompt = {}
        for label, model_output in [("prompt1", out1), ("prompt2", out2), ("prompt3", out3)]:
            prompt = generate_evaluation_prompt(product_facts, model_output)
            print(f"Bewerte Produkt {pid} f√ºr {label}...")
            try:
                score_str = run_evaluation_prompt(prompt)
                scores = [int(s.strip()) for s in score_str.split(",")]
            except Exception as e:
                print(f"Fehler bei Bewertung von Produkt {pid} ({label}): {e}")
                scores = [None]*5
            scores_per_prompt[label] = scores
            time.sleep(1.5)  # Rate limiting

        # Gesamtwertung: Summe der 5 Kriterien (je h√∂her besser)
        total_scores = {k: sum(v) if all(v) else 0 for k,v in scores_per_prompt.items()}

        # Ranking der Prompts (1=best, 3=schlechteste)
        ranked = sorted(total_scores.items(), key=lambda x: x[1], reverse=True)
        ranking = {ranked[i][0]: i+1 for i in range(len(ranked))}

        all_results.append({
            "ProductId": pid,
            "prompt1_scores": scores_per_prompt["prompt1"],
            "prompt2_scores": scores_per_prompt["prompt2"],
            "prompt3_scores": scores_per_prompt["prompt3"],
            "prompt1_total": total_scores["prompt1"],
            "prompt2_total": total_scores["prompt2"],
            "prompt3_total": total_scores["prompt3"],
            "prompt1_rank": ranking["prompt1"],
            "prompt2_rank": ranking["prompt2"],
            "prompt3_rank": ranking["prompt3"]
        })

    # DataFrame aus Ergebnissen bauen
    results_df = pd.DataFrame(all_results)

    # Durchschnittliche Scores und Ranks pro Prompt berechnen
    avg_scores = {
        "prompt1": results_df["prompt1_total"].mean(),
        "prompt2": results_df["prompt2_total"].mean(),
        "prompt3": results_df["prompt3_total"].mean()
    }
    avg_ranks = {
        "prompt1": results_df["prompt1_rank"].mean(),
        "prompt2": results_df["prompt2_rank"].mean(),
        "prompt3": results_df["prompt3_rank"].mean()
    }

    print("\nDurchschnittliche Gesamtscores pro Prompt:")
    for k,v in avg_scores.items():
        print(f"{k}: {v:.2f}")

    print("\nDurchschnittliche Rangpl√§tze pro Prompt (niedriger = besser):")
    for k,v in avg_ranks.items():
        print(f"{k}: {v:.2f}")

    # Ergebnisse speichern
    results_df.to_csv("vergleich_prompt1_2_3_bewertung.csv", index=False)
    print("\n‚úÖ Vergleichsbewertung gespeichert in 'vergleich_prompt1_2_3_bewertung.csv'.")

if __name__ == "__main__":
    main()
```
